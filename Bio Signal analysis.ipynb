```python
#Removing outliers

#lower limit Q1-1.5*IQR
#upper limit Q3+1.5*IQR


cols=['age','height(cm)', 'weight(kg)', 'waist(cm)','systolic', 'relaxation', 'fasting blood sugar', 'Cholesterol',
       'triglyceride', 'HDL', 'LDL', 'hemoglobin','serum creatinine', 'AST', 'ALT', 'Gtp']

for i in cols:
    Q1=df[i].quantile(0.25)
    Q3=df[i].quantile(0.75)

    IQR=Q3-Q1
    df=df[((df[i]>=Q1-1.5*IQR) & (df[i]<=Q3+1.5*IQR))]
```

```python
for i in col:
    if (df[i].dtypes=='int64' or df[i].dtypes=='float64'):
        sns.boxplot(df[i])
        plt.xlabel(i)
        plt.ylabel('count')
        plt.show()
```

```python
#Feature selection 

x=df.iloc[:,:-1]
y=df['smoking']

from sklearn.ensemble import ExtraTreesClassifier
model=ExtraTreesClassifier()
model.fit(x,y)
df_new=pd.Series(model.feature_importances_,index=x.columns)
plt.figure(figsize=(9,9))
df_new.nlargest(df.shape[1]).plot(kind='barh')
plt.show()
```

```python
#selecting top 15 columns according

columns=['gender', 'Gtp','triglyceride', 'age', 'height(cm)', 'weight(kg)', 'hemoglobin','waist(cm)','HDL',
         'fasting blood sugar','ALT','LDL','systolic','relaxation','serum creatinine',]

x=df[columns]
y=df['smoking']
```

```python
#Dividing data set into train and test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
#Scaling data

from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
#Model building
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
```

```python
models =[("LogisticReg", LogisticRegression(max_iter=1000)),("SVC", SVC()),
("DecisionTree", DecisionTreeClassifier()),("NaiveBayes", GaussianNB()),
('RandomForest',RandomForestClassifier()),('GradientBoost',GradientBoostingClassifier())]
```

```python
#Calculating accuracy of different models
names = []
accuracies = []
finalResults = []
for name,model in models:
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    accuracyScore = accuracy_score(y_test, y_pred)
    accuracies.append(accuracyScore)
    names.append(name)
    finalResults.append((name,accuracyScore))
```

```python
finalResults_df = pd.DataFrame({'Model-Name': names, 'Accuracy-Score': accuracies})
finalResults_df.sort_values(by='Accuracy-Score', ascending=False)
```

```python
#selecting random forest classifier as final model
rf = RandomForestClassifier(n_estimators = 1000)
rf.fit(x_train, y_train)
```

```python
y_pred = rf.predict(x_test)
accuracy_score(y_test, y_pred)
```
